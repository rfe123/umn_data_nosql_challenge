### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print(f'There are {establishments.count_documents(query)} documents that match the query.')

results = list(establishments.find(query))

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f"There are {len(df.values)} rows in the Data Frame.")
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex':'London'}, 'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
print(f'There are {establishments.count_documents(query)} documents that match the query.')
results = list(establishments.find(query))

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f"There are {len(df.values)} rows in the Data Frame.")
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

#Find the new restaurant by name
new_restaurant = establishments.find_one({"BusinessName": "Penang Flavours"})

#store coordinates of this restaurant
degree_search = 0.01
latitude = new_restaurant['geocode']['latitude']
longitude = new_restaurant['geocode']['longitude']

query = {'RatingValue': 5, 'geocode.latitude': {'$gte': latitude-degree_search, '$lte': latitude+degree_search}, 'geocode.longitude': {'$gte': longitude-degree_search, '$lte': longitude+degree_search}}
sort =  [('scores.Hygiene',1)]

# Print the results
results = list(establishments.find(query).sort(sort))
print(f"{new_restaurant['BusinessName']} Latitude:{latitude} Longitude:{longitude}\n---------------------------")
for result in results[:5]:
    print(f"[{result['BusinessName']}] Hygiene Score:{result['scores']['Hygiene']} Latitude:{result['geocode']['latitude']} Longitude:{result['geocode']['longitude']}")
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(pd.json_normalize(results))
df[['BusinessName', 'scores.Hygiene', 'geocode.latitude', 'geocode.longitude']].head()
```
